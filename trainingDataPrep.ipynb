```python
import pandas as pd
from pylab import *
```

```python
df = pd.read_csv('data.csv')
df
```

```python
# Drop unwanted columns
data = df.drop(columns=['Name','Equipment','Age','BodyweightKg','Squat1Kg','Squat2Kg','Squat3Kg','Squat4Kg',
                        'Bench1Kg','Bench2Kg','Bench3Kg','Bench4Kg',
                        'Deadlift1Kg','Deadlift2Kg','Deadlift3Kg','Deadlift4Kg',
                        'TotalKg','Place','Wilks','McCulloch','Glossbrenner','Tested','Date'])
data
```

```python
# Dropping lifters with negative Best values (arises from people placing DQ)
data.drop(data[data['Best3SquatKg']< 0].index, inplace=True)
data.drop(data[data['Best3BenchKg']< 0].index, inplace=True)
data.drop(data[data['Best3DeadliftKg']< 0].index, inplace=True)
# Dropping lifters who have any missing Best values
data.dropna(subset=['Best3SquatKg','Best3BenchKg','Best3DeadliftKg'], inplace=True)
```

```python
# Discretising each Best Kg column
S_conditions = [data['Best3SquatKg'] < 100,
                data['Best3SquatKg'] < 190,
                data['Best3SquatKg'] < 380,
                data['Best3SquatKg'] >=380]
S_choices = ['0to99','100to189','190to379','380+']
data['Best3SquatKg'] = np.select(S_conditions, S_choices, default='380+')

B_conditions = [data['Best3BenchKg'] < 70,
                data['Best3BenchKg'] < 125,
                data['Best3BenchKg'] < 280,
                data['Best3BenchKg'] >=280,]
B_choices = ['0to69','70to124','125to279','280+']
data['Best3BenchKg'] = np.select(B_conditions, B_choices, default='280+')

D_conditions = [data['Best3DeadliftKg'] < 120,
                data['Best3DeadliftKg'] < 215,
                data['Best3DeadliftKg'] < 340,
                data['Best3DeadliftKg'] >=340]
D_choices = ['0to119','120to214','215to339','340+']
data['Best3DeadliftKg'] = np.select(D_conditions, D_choices, default='340+')
```

```python
data.to_csv('trainingData.csv', index=False)
```
