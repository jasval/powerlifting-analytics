## Importing Dataset from Kaggle and Official OpenPowerlifting site

Alongside importing data we proceed to convert 'Date' to date-dtype, to cut and merge the dataset based on the timeseries. </br>
Additionally, we will define constants to be used throughout the cleaning process.

```python
df = pd.read_csv('openpowerlifting.csv', sep=',', low_memory=False)
df.drop(['MeetName', 'MeetState', 'MeetCountry', 'Event', 'Division', 'IPFPoints', 'Country', 'Federation'], axis=1, inplace=True)
df2 = pd.read_csv('openpowerlifting-latest.csv', sep=',', low_memory=False)
df2.drop(['BirthYearClass', 'ParentFederation', 'MeetName', 'MeetState', 'MeetCountry', 'Event', 'Division', 'IPFPoints', 'Country', 'Federation'], axis=1, inplace=True)

cols = ['Squat1Kg','Squat2Kg','Squat3Kg','Squat4Kg','Best3SquatKg','Bench1Kg','Bench2Kg','Bench3Kg','Bench4Kg','Best3BenchKg','Deadlift1Kg','Deadlift2Kg','Deadlift3Kg','Deadlift4Kg','Best3DeadliftKg']
df['Date'] = pd.to_datetime(df2['Date'], format='%Y-%m-%d')
df2['Date'] = pd.to_datetime(df2['Date'], format='%Y-%m-%d')

date_limit = pd.to_datetime('2019-04-20', format='%Y-%m-%d')
df2.drop(df2[df2['Date'] <= date_limit].index, inplace=True)

df.append(df2)
```

## Cleaning the Data

```python
df.drop_duplicates(inplace=True)
df[cols] = df[cols].clip(lower=0)
df[cols] = df[cols].fillna(0)
df.dropna(subset=['Age', 'Name'], inplace=True)
df.drop(df[df['Age'] < 18].index, inplace=True)
df.drop(df[df['Sex'] == "Mx"].index, inplace=True)
df.dropna(subset=['BodyweightKg'], axis=0, inplace=True)
df.drop(df[df['BodyweightKg']< 30].index, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

## Discretisation and Binning

```python
conditions = [
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 43),
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 47),
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 52),
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 57),
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 63),
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 72),
    (df['Sex'] == 'F') & (df['BodyweightKg'] < 84),
    (df['Sex'] == 'F'),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 53),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 59),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 66),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 74),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 83),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 93),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 105),
    (df['Sex'] == 'M') & (df['BodyweightKg'] < 120),
    (df['Sex'] == 'M')]
choices = ['43','47','52','57','63','72','84','84+',
          '53','59','66','74','83','93','105','120','120+']
df['WeightClassKg'] = np.select(conditions, choices, default='120+')
```
